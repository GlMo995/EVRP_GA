```python
import random
from matplotlib import pyplot as plt
import numpy as np
```

```python
max_x = 1000
max_y = 1000

num_clients = 5
num_vehicles = 2

random.seed(3)
exp_config = {
    "num_clients": num_clients,
    "num_vehicles": num_vehicles,
    "base_coord": np.array([random.randint(0, max_x), random.randint(0, max_y)]).astype(np.float32),
    "clienst_coord": np.array([(random.randint(0, max_x), random.randint(0, max_y)) for i in range(num_clients)]).astype(np.float32),
    "vehicles_init_soc": [1]*(num_vehicles), 
    "vehicles_battery_capacity": np.array([4]*(num_vehicles)),  # in Ah
    "vehicles_weight": np.array([40]*(num_vehicles)),           # in kg
    "dt": 3, # in seconds
    
    "vehicles_max_payload": np.array([4]*(num_vehicles)),       # in kg
    "max_speed": 10,        # in m/s
    "max_acc": 5,           # in m/s2
}
```

```python
def plot_map(exp_config):
    fig, axs = plt.subplots(nrows=1, ncols=1, figsize=(4,4))
    ax = axs
    ax.scatter(exp_config["base_coord"][0], exp_config["base_coord"][1], color="green")
    ax.scatter(exp_config["clienst_coord"][:,0], exp_config["clienst_coord"][:,1])
    ax.set_xlabel("x (m)")
    ax.set_ylabel("y (m)")

plot_map(exp_config)
```
